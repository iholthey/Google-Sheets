```python
# 1. Authentifizierung (Du musst den Zugriff in einem Popup-Fenster erlauben)
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
import pandas as pd

# Anmeldedaten für Google Sheets bereitstellen
creds, _ = default()
gc = gspread.authorize(creds)

# 2. Das Spreadsheet über die URL öffnen
spreadsheet_url = 'https://docs.google.com/spreadsheets/d/1umVwJVKR2zc8JTjzFk_03IJNuhfnZnWXeixmpw1Ma6c/edit?usp=drivesdk'
sh = gc.open_by_url(spreadsheet_url)

# 3. Das Tabellenblatt "Leistungsnachweis" (Lasche) auswählen
# Hinweis: Achte darauf, dass der Name exakt so geschrieben ist wie im Sheet
worksheet = sh.worksheet("Leistungsnachweis")

# 4. Alle Daten auslesen
rows = worksheet.get_all_values()

# 5. In ein Pandas DataFrame umwandeln (macht die Auswertung einfacher)
# Die erste Zeile wird als Spaltenüberschrift verwendet

# Function to make column names unique
def make_columns_unique(columns):
    seen = {}
    unique_columns = []
    for col in columns:
        if col in seen:
            seen[col] += 1
            unique_columns.append(f"{col}_{seen[col]}")
        else:
            seen[col] = 0
            unique_columns.append(col)
    return unique_columns

# Apply the function to the header row
unique_headers = make_columns_unique(rows[0])

df = pd.DataFrame(rows[1:], columns=unique_headers)

# 6. Ergebnis anzeigen
print("Daten erfolgreich geladen!")
print(f"Anzahl der Zeilen: {len(df)}")
print("-" * 30)

# Zeigt die ersten 10 Zeilen und alle Spalten an
df.head(10)
```
